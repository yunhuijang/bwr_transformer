```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Advisen_long_lat.csv')
```

```python
freq = data.groupby(['Year', 'CASE_TYPE'], as_index=False).size()
freq['Year'] = freq['Year'].astype(int)
```

```python
pivotted_freq = freq.pivot(index='Year', columns='CASE_TYPE', values='size').fillna(0)
```

```python
normalized_df = pivotted_freq.T/pivotted_freq.T.sum(axis=0)
```

```python
columns = normalized_df.sum(axis=1).sort_values(ascending=False).index.tolist()
```

```python
final_df = normalized_df.T[columns]
```

```python
# add others
top_columns = normalized_df.sum(axis=1).sort_values(ascending=False)[:8].index.tolist()
top_columns.extend(['others' for i in range(len(columns)-len(top_columns))])
final_df_with_others = final_df.copy()
final_df_with_others.columns = top_columns
final_df_with_others['others_sum'] = final_df_with_others.loc[:,'others'].sum(axis=1)
final_df_with_others = final_df_with_others.drop(['others'], axis=1)
```

```python
# without others
fig, ax = plt.subplots(figsize=(10,10))
final_df.plot(kind='bar', stacked=True, ax=ax)
ax.legend(bbox_to_anchor = (1.5,1))
```

```python
# with others
fig, ax = plt.subplots(figsize=(10,10))
final_df_with_others.plot(kind='bar', stacked=True, ax=ax)
ax.legend(bbox_to_anchor = (1.5,1))
```
